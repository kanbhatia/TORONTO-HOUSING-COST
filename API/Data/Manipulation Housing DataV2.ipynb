```python
# IMPORT LIBRARIES

%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import datetime as dt

from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
dateparse = lambda x: pd.datetime.strptime(x, "%Y-%m-%d")
```

```python
# READ CSV WITH DEVELOPMENT APPLICATIONS FROM TORONTO DATASET = https://open.toronto.ca/dataset/development-applications/

data = pd.read_csv("Development_Applications_Data.csv", parse_dates=['DATE_SUBMITTED'], date_parser=dateparse)
data
```

```python
# PARSE THROUGH DATE TO GET DAY, MONTH AND YEAR SEPARATED 

data['DATE_SUBMITTED'] = pd.to_datetime(data['DATE_SUBMITTED'])
data['YEAR'], data['MONTH'], data["DAY"] = data['DATE_SUBMITTED'].dt.year, data['DATE_SUBMITTED'].dt.month, data['DATE_SUBMITTED'].dt.day
data.head()
```

```python
# TAKE ONLY CONDO DEVELOPMENTS FROM THE DATA

grouped_data = data.loc[data["APPLICATION_TYPE"] == "CD"]
grouped_data
```

```python
# REMOVE NOT NECESSARY COLUMNS FROM THE DATAFRAME

filtered_data = grouped_data[["APPLICATION_TYPE", "YEAR", "STATUS",
                           "STREET_NAME", "STREET_NUM", "STREET_TYPE", "POSTAL","_id"]]
filtered_data.sort_values("YEAR").head()
```

```python
# GROUP THE ADDRESS COLUMNS TOGETHER TO GET ONE STRING VALUE FOR THE ADDRESS

filtered_data['ADDRESS'] = filtered_data[['STREET_NUM', 'STREET_NAME', 'STREET_TYPE']].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
filtered_2 = filtered_data[["APPLICATION_TYPE", "YEAR", "STATUS",
                           "ADDRESS", "POSTAL", "_id"]].sort_values("YEAR")
filtered_2
```

```python
# READ RENTAL PRICES CSV FROM CANADA MORTGAGE AND HOUSING CORPORATION
#https://www03.cmhc-schl.gc.ca/hmip-pimh/en/TableMapChart/Table?TableId=2.2.11&GeographyId=2270&GeographyTypeId=3&DisplayAs=Table&GeograghyName=Toronto

rent = pd.read_csv("rental_prices.csv")
rent.head()
```

```python
# FILTER THE DATA FOR NEEDED COLUMNS ONLY

rent_filtered = rent[["Year","Bachelor", "1 Bedroom", "2 Bedroom", "3 Bedroom +", "Total"]]

rent_filtered = rent_filtered.rename(columns={"Year": "YEAR"})
rent_filtered.head()
```

```python
# MERGE THE FILTERED RENT AND APPLICATION DATA TOGETHER ON YEAR

rent_application = pd.merge(rent_filtered, filtered_2[["APPLICATION_TYPE", "STATUS",
                           "ADDRESS", "POSTAL", "_id", "YEAR"]],on='YEAR',how='right')
rent_application

# RENAME A COLUMN

rent_application = rent_application.rename(columns={"POSTAL": "Postcode"})
rent_application.head()

# REMOVE 2019
rent_application = rent_application[rent_application.YEAR != 2019]

# EXPORT TO CSV

rent_application.to_csv("rent_application_per_condo_size.csv")
rent_application
```

```python
# READ POSTAL CODE DATA WITH NEIGHBORHOODS

postal_codes_1 = pd.read_csv("postal_codes.csv")

postal_codes = postal_codes_1[["Postcode", "Borough"]]
df = postal_codes.drop_duplicates()
df.head()
```

```python
# MERGE THE RENT APPLICATION DATA TO THE POSTCODE DATA ON POSTCODE

rent_final = pd.merge(rent_application, df[["Postcode"]],on='Postcode',how='inner')

rent = rent_final.sort_values("_id")

rent_final = rent_final.rename(columns={"YEAR": "Year"})

rent_final.head()
```

```python
# READ CSV WITH RENTS PER NEIGHBORHOOD

rent_neighborhoods = pd.read_csv("rent_neighborhoods.csv")
rent_neighborhoods.head()
```

```python
# MERGE NEIGHBORGHOOD RENTS WITH POSTAL CODES ON NEIGHBORHOOD TO GET 
# POSTAL CODES PER LINES TO MERGE WITH APPLICATION DATA
rent_merged = pd.merge(rent_neighborhoods, postal_codes_1[["Neighbourhood", "Postcode"]],on='Neighbourhood',how='inner')

rent_merged.head()
```

```python
rent_appli_neigh = pd.merge(rent_final, rent_merged[["Neighbourhood", "Rent", "Year", "Postcode"]], on=['Year','Postcode'], how='left')
rent_appli_neigh_clean = pd.merge(rent_final, rent_merged[["Neighbourhood", "Rent", "Year", "Postcode"]], on=['Year','Postcode'], how='inner')

final_table_all_values = rent_appli_neigh.sort_values("_id")
final_table_all_values
```

```python
final_table_clean = rent_appli_neigh_clean.sort_values("Year")
final_table_clean
```

```python

final_table_all_values.to_csv("final_table_all_values.csv")
final_table_clean.to_csv("final_table_clean.csv")
```
